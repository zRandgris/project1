```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt 
import numpy as np

```

```python
csv_path = Path("Resource/Data1.csv")
df = pd.read_csv(csv_path)
df.head(50)
```

```python
# Assume df is your DataFrame and 'date' is your column with dates
df['REF_DATE'] = pd.to_datetime(df['REF_DATE'], format='%Y-%m')

# Set 'date' as index if it's not yet
df.set_index('REF_DATE', inplace=True)

# Resample to yearly data, taking the mean of each year
df_yearly = df.resample('Y').mean()


df_yearly
```

```python
clean_df = df_yearly[["VALUE"]]

clean_df
```

```python
# Pro #1 The Olympics increase valuable tourism, which can boost local economies.
categories = ['Food', 'Alcoholic beverages', 'Clothing and footwear', 'Electricity Gas and other fuels', 'Transport', 'Transport services', 'Newspapers Books and stationery', 'Food and beverage services']

values1 = [49947, 13103, 15981, 22131, 57294, 10342, 6611, 34010]
values2 = [51852, 13216,16957, 24087, 62486, 10523, 6994, 35774]

# Set the position of each bar on the x-axis
x = np.arange(len(categories))

# Set the width of the bars
width = 0.5

# Create the figure and axis objects
fig, ax = plt.subplots()

# Plot the first set of bars
rects1 = ax.bar(x - width/2, values1, width, label='Q1')

# Plot the second set of bars
rects2 = ax.bar(x + width/2, values2, width, label='Q4')

# Customize the appearance
ax.set_title('Household Final Consumption For The Year 1976')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()
plt.xticks(rotation=90)

# Add labels on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

# Display the plot
plt.show()
```
